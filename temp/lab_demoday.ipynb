```python
from dartfx.postman import mtnards, postman, socrata
import os
import logging
logging.basicConfig(level=logging.DEBUG)
```

```python
# Initialize environment
rds_server = mtnards.MtnaRdsServer()
rds_server.host = "https://www172.highvaluedata.net"
rds_server.api_key = os.getenv("MTNA_RDS_API_KEY")
rds_server.ssl_verify = False

postman_api = postman.PostmanApi(os.getenv("POSTMAN_API_KEY"))
host_workspace_id = os.getenv("POSTMAN_WS_DEMO_DAY")
publisher= mtnards.MtnaRdsPostmanPublisher(postman_api,rds_server, temp_workspace_id=host_workspace_id)
```

```python
# Create a Data Product Collection
collection = postman.DataProductCollectionManager.create(postman_api, host_workspace_id, '🔢 American National Election Study 1948')
metadata_folder_id = collection.create_folder("ℹ Metadata")
basex_folder_id = collection.create_folder("⛃ BaseX", parent_id=metadata_folder_id)
vos_folder_id = collection.create_folder("⛃ Virtuoso", parent_id=metadata_folder_id)
mtnards_folder_id = collection.create_folder("⛁ MTNA RDS")
dreamfactory_folder_id = collection.create_folder("⛁ Dreamfactory")
postgrest_folder_id = collection.create_folder("⛁ PostgREST")
hasura_folder_id = collection.create_folder("⛁ Hasura")
hasura_folder_id = collection.create_folder("⛁ Postgraphile")
```

```python
# RDS Data product to collection sub-folder
host_collection_id = collection.id
host_folder_id = mtnards_folder_id
folder_id = publisher.publish_data_product_to_collection('us_anes','anes1948', host_collection_id, host_folder_id, create_root=False)
```
